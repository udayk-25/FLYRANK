## 3. Quick look at the data (2-3 real numbers)

To support this lane choice, we analyze the 30,000-row anonymized starter dataset (`data/raw/content_refresh_anonymized.csv`). Below, we load the dataset and compute three key metrics:
1. **The Overall Decline Rate**: The baseline rate of pages labeled as declining (`trend_direction == 'down'`).
2. **Decline Rate among High-Visibility Pages**: The rate of decline among pages that actually get significant search exposure (`impressions_90d >= 500`).
3. **The CTR Gap**: A comparison of click-through rates (CTR) between declining and stable pages, showcasing the decay of traffic capture.


```python
import pandas as pd
import numpy as np

# Load the starter CSV
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# 1. Overall Decline Rate
total_pages = len(df)
declining_pages = df[df["trend_direction"].str.lower().eq("down")].shape[0]
decline_rate = declining_pages / total_pages
print(f"Total pages in starter dataset: {total_pages:,}")
print(f"Overall decline rate (trend_direction == 'down'): {decline_rate:.1%} ({declining_pages:,} pages)")

# 2. Decline Rate among High-Visibility Pages (impressions_90d >= 500)
visible_df = df[df["impressions_90d"] >= 500]
total_visible = len(visible_df)
declining_visible = visible_df[visible_df["trend_direction"].str.lower().eq("down")].shape[0]
visible_decline_rate = declining_visible / total_visible
print(f"\nTotal high-visibility pages (impressions_90d >= 500): {total_visible:,}")
print(f"Decline rate among high-visibility pages: {visible_decline_rate:.1%} ({declining_visible:,} pages)")

# 3. Median CTR Comparison: Declining vs. Stable Pages
# Note: CTR in this dataset is represented as a percentage (e.g. 0.76 = 0.76%, not 76%)
median_ctr_declining = df[df["trend_direction"].str.lower().eq("down")]["ctr"].median()
median_ctr_stable = df[df["trend_direction"].str.lower().eq("stable")]["ctr"].median()
print(f"\nMedian CTR for declining pages: {median_ctr_declining:.3f}%")
print(f"Median CTR for stable pages: {median_ctr_stable:.3f}%")
print(f"Stable pages capture {median_ctr_stable / (median_ctr_declining + 1e-9):.1f}x higher CTR than declining pages.")
```

- **Observed performance**: I can report historical, observed metrics from the dataset showing differences between declining and stable content.

### What I CANNOT claim (causal proof / 'predicting Google'):
- **Causal proof**: I cannot claim that updating a page will *cause* it to recover or grow in search rankings. The dataset does not contain random assignments or experimental controls to prove causation.
- **Predicting Google's algorithm**: I cannot claim to have reverse-engineered or predicted the internal ranking algorithms of Google or other search engines. The model merely observes how historical page performance changes over time.
- **Exact traffic forecasts**: I cannot claim to predict the exact number of clicks or impressions a page will get in the future, only the directional probability of decline.
